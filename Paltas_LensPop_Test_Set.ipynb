```python
N_rows = 8
fig,ax = pl.subplots(2,N_rows,figsize=(18,6))
fig2,ax2 = pl.subplots(2,N_rows,figsize=(18,6))
fig.suptitle('SNR_i > 20\n',fontsize=25,fontweight='bold')
fig2.suptitle('SNR_i < 20\n',fontsize=25,fontweight='bold')
n_pass = 0
n_fail = 0
metadata_dict = {}
metadata_dict['with_RSP'] =  pd.read_csv('/mnt/zfsusers/hollowayp/paltas/mag_npy_files/Paltas_with_RSP/metadata.csv')
metadata_dict['no_RSP'] = pd.read_csv('/mnt/zfsusers/hollowayp/paltas/mag_npy_files/Paltas_no_RSP/metadata.csv')
metadata_dict['with_RSP_no_sub'] = pd.read_csv('/mnt/zfsusers/hollowayp/paltas/mag_npy_files/Paltas_with_RSP_no_sub/metadata.csv')
metadata_dict['no_RSP_no_sub'] = pd.read_csv('/mnt/zfsusers/hollowayp/paltas/mag_npy_files/Paltas_no_RSP_no_sub/metadata.csv')

vmin=None;vmax=None
stretch = 'linear';interval='zscale'
for metadata_indx in range(1000):
    skip=False
    LensPop_indx = metadata_dict['with_RSP'].loc[metadata_indx]['Catalogue_Index']
    metadata_indx_dict = {};Paltas_img_dict = {}
    for criterion in ['with_RSP','no_RSP','with_RSP_no_sub','no_RSP_no_sub']:
        metadata_indx_dict[criterion] = np.where(metadata_dict[criterion]['Catalogue_Index']==LensPop_indx)[0]
        if len(metadata_indx_dict[criterion])==0: skip=True; continue
        else: metadata_indx_dict[criterion]=int(metadata_indx_dict[criterion])
        Paltas_img_dict[criterion] = np.load(f'/mnt/zfsusers/hollowayp/paltas/mag_npy_files/Paltas_{criterion}'+\
                                             f'/image_{str(metadata_indx_dict[criterion]).zfill(7)}.npy')
    if skip: continue
    SNR_i_band = np.round(db_LensPop_LSST.loc[LensPop_indx]['i_see_sub'],1)
    Paltas_magnification = np.round(metadata_dict['with_RSP'].loc[metadata_indx]['Magnification'],1)
    if SNR_i_band>20 and n_pass<N_rows:
        for p_i,image_i in enumerate([Paltas_img_dict['with_RSP'],Paltas_img_dict['no_RSP']]):
                                    #  Paltas_img_dict['with_RSP_no_sub'],Paltas_img_dict['no_RSP_no_sub']]):
            norm = ImageNormalize(image_i, interval=norm_dict[interval](), stretch=norm_dict[stretch]())
            ax[p_i,n_pass].imshow(image_i,norm=norm,cmap='Greys_r')
        # ax[0,n_pass].imshow(Paltas_img_dict['with_RSP'],
        #     vmin = np.percentile(Paltas_img_dict['with_RSP'],20), vmax=np.percentile(Paltas_img_dict['with_RSP'],95))
        # ax[1,n_pass].imshow(Paltas_img_dict['no_RSP'],vmin=vmin,vmax=vmax)
        # ax[2,n_pass].imshow(np.log10(Paltas_img_dict['with_RSP_no_sub']))
        # ax[3,n_pass].imshow(np.log10(Paltas_img_dict['no_RSP_no_sub']))
        ax[0,n_pass].set_title(f'SNR: {SNR_i_band}, Mag: {Paltas_magnification}')
        for i in range(2):
          ax[i,n_pass].set_xticks([]);ax[i,n_pass].set_yticks([])
        n_pass+=1
    if SNR_i_band<20 and n_fail<N_rows:
        for p_i,image_i in enumerate([Paltas_img_dict['with_RSP'],Paltas_img_dict['no_RSP']]):
            norm = ImageNormalize(image_i, interval=norm_dict[interval](), stretch=norm_dict[stretch]())
            ax2[p_i,n_fail].imshow(image_i,norm=norm,cmap='Greys_r')
        # ax2[0,n_fail].imshow(Paltas_img_dict['with_RSP'],vmin=vmin,vmax=vmax)
        # ax2[1,n_fail].imshow(Paltas_img_dict['no_RSP'],vmin=vmin,vmax=vmax)
        # ax2[2,n_fail].imshow(np.log10(Paltas_img_dict['with_RSP_no_sub']))
        # ax2[3,n_fail].imshow(np.log10(Paltas_img_dict['no_RSP_no_sub']))
        ax2[0,n_fail].set_title(f'SNR: {SNR_i_band}, Magnif: {Paltas_magnification}')
        for i in range(2):
            ax2[i,n_fail].set_xticks([]);ax2[i,n_fail].set_yticks([])
        n_fail+=1
    if n_fail==N_rows and n_pass==N_rows: break
fig.tight_layout()
fig2.tight_layout()
pl.show()
```
